```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
def f(x, yi1, yi, xi1, xi):
    return ((yi1 - yi)/(xi1 - xi)) * (x - xi) + yi
    
n = int(input())
x = sorted(np.array([random.uniform(0, n) for i in range(n)]))
y = np.random.uniform(-n, n, n)
print(x)
print(y)
plt.plot(x, y)
plt.show
m = int(input())
z = sorted(np.array([random.uniform(x[0], x[n - 1]) for i in range(m)]))
ans = np.zeros(m)
print(z)
for i in range(m):
    for j in range(n):
        if z[i] < x[j]:
            k = j
            break
    ans[i] = f(z[i], y[k], y[k - 1], x[k], x[k - 1])
print(ans)
```
