```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
def f(x, yi1, yi, xi1, xi):
    return ((yi1 - yi)/(xi1 - xi)) * (x - xi) + yi
    
def read_data(filename):
    file = open(filename, 'r')
    inputData = [float(line) for line in file]
    file.close()
    return inputData

def write_data(filename, outputVec):
    file = open(filename, 'w')
    for ans in outputVec:
        file.write(str(ans) + '\n')
    file.close()
    return

x = np.array(read_data("train/train.dat"))
y = np.array(read_data("train/train.ans"))
z = np.array(read_data("test/test.dat"))
m = len(z)
n = len(x)
ans = np.zeros(m)
for i in range(m):
    for j in range(n):
        if z[i] < x[j]:
            k = j
            break
    ans[i] = f(z[i], y[k], y[k - 1], x[k], x[k - 1])
write_data("test/test.ans", ans)
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(z, ans)
ax.plot(x, y)
fig.savefig("linear_interpolation.png")
```
